**NumPy Kütüphanesi: Temel Kavramlar ve Önemi**

NumPy (Numerical Python), Python programlama dili için geliştirilmiş, bilimsel hesaplamalar ve çok boyutlu verilerle çalışmak için kullanılan en temel kütüphanelerden biridir. Makine öğrenmesi algoritmalarının büyük çoğunluğu doğrusal cebire dayandığı için NumPy, bu alandaki veri setlerinin temsil edilmesi ve işlenmesinde kritik bir rol oynar.

**1. Vektörler, Matrisler ve Doğrusal Cebir İlişkisi**

Makine öğrenmesindeki veriler matematiksel olarak vektörler ve matrisler ile ifade edilir:

**Vektör:** Bir büyüklük ve yön ile tanımlanan niteliktir. Boyut sayısına göre (n-elemanlı) düzlemde veya çok boyutlu uzayda bir nokta ya da ok olarak temsil edilir.

**Matris:** Sayıların satırlar ve sütunlar halinde düzenlendiği tablo yapısıdır.

**Doğrusal Cebir:** Veri setlerindeki özniteliklerin koordinat düzleminde veya uzayda döndürülmesi, ölçeklenmesi ve hareket ettirilmesi gibi dönüşümleri inceleyen matematik dalıdır.

**2. NumPy Dizileri (Arrays)**

NumPy'ın en temel veri yapısı n-boyutlu dizilerdir. Python'un standart listelerinden farklı olarak, bu diziler matematiksel işlemler için optimize edilmiştir:

**Vektör Temsili:** Tek boyutlu bir NumPy dizisi bir vektörü temsil eder.

**Matris Temsili:** İki boyutlu bir NumPy dizisi bir matrisi temsil eder.

n-Boyutlu Uzaylar: Daha karmaşık veri yapıları (örneğin renkli görüntüler) çok boyutlu diziler ile ifade edilir.

**3. Temel NumPy Fonksiyonları ve Kullanımı**


**Kütüphane Tanımlama:** Genellikle import numpy as np şeklinde koda dahil edilir.

**Dizi Oluşturma:** np.array([1, 2, 3]): Belirli değerlerden dizi oluşturur.

np.zeros((satir, sutun)): Tamamı sıfırlardan oluşan bir matris oluşturur.

np.ones((satir, sutun)): Tamamı birlerden oluşan bir matris oluşturur.

np.arange(baslangic, bitis, artis): Belirli bir aralıkta sıralı sayılar üretir.

**Dizi Özelliklerini Sorgulama:**

ndim: Dizinin boyut sayısını verir.

shape: Dizinin satır ve sütun bilgisini (boyutlarını) verir.

size: Dizideki toplam eleman sayısını verir.

dtype: Dizideki verilerin tipini (float64, int32 vb.) gösterir.

1. NumPy Kütüphanesini Ekleme ve İlk Vektör
Bu kısım, kütüphaneyi içeri aktarır ve İris veri setinden bir örneği vektör olarak tanımlar.

```python
import numpy as np

# İris veri setine ait bir örneğin vektör olarak tanımlanması
# Öznitelikler: taç yaprak genişliği, taç yaprak uzunluğu, çanak yaprak genişliği, çanak yaprak uzunluğu
iris = np.array([5.7, 2.6, 3.5, 1.0])
print("İris Vektörü:", iris)
```

2. Dizi (Matris) Oluşturma Yöntemleri
Farklı boyutlarda ve başlangıç değerlerinde (sıfır, bir, sabit değer) diziler oluşturma örnekleridir.

```python
# Bir ve iki boyutlu diziler oluşturma
print("Bir Boyutlu Dizi:", np.array([[2, 5, 7]]))
print("İki Boyutlu Dizi:\n", np.array([[2, 5, 7], [3, 6, 8]]))
```

```python
# Sıfırlardan oluşan diziler (Zeros)
print("3 Elemanlı Sıfır Dizisi:", np.zeros(3))
print("4x3'lük Sıfır Matrisi:\n", np.zeros((4, 3)))
print("3 Boyutlu (2x4x3) Sıfır Dizisi:\n", np.zeros((2, 4, 3)))
```

```python
# Birlerden oluşan dizi (Ones)
print("4 Elemanlı Birler Dizisi:", np.ones(4))
```

```python

# Belirli bir değerle doldurulmuş dizi (Full)
print("Metin İçeren Dizi:\n", np.full((2, 2), "Python"))
```

3. Sayı Aralıkları ve Rastgele Sayılar
İstatistikte ve veri hazırlamada sıkça kullanılan sayı üretim fonksiyonları.

```python
# Belirli aralıklarda sayı dizileri oluşturma
print("Arange (0-1 arası 0.2 artış):", np.arange(0, 1, 0.2))
print("Linspace (0-1 arasını 5'e böl):", np.linspace(0.0, 1.0, num=5))
print("Linspace (Bitiş değeri hariç):", np.linspace(0.0, 1.0, num=5, endpoint=False))
```

```python
# Rastgele sayı üretimi
print("Rastgele (0-1 arası):\n", np.random.rand(2, 2))
print("\nNormal Dağılımlı Rastgele Sayılar:\n", np.random.randn(2, 2))
print("\nBelirli Aralıkta (3-7) Rastgele Tam Sayılar:\n", np.random.randint(3, 7, size=(2, 4)))
```

4. Dizi Özellikleri ve Şekillendirme
Dizinin tipini öğrenme, boyutlarını değiştirme ve düzleştirme işlemleri.

```python
# Veri tipi ve eleman boyutu
dizi = np.arange(2.0, 9.0)
print("Veri Tipi:", dizi.dtype)

dizi_64bit = np.arange(1, 3, dtype=np.int64)
print("Eleman Boyutu (Byte):", dizi_64bit.itemsize)
```

```python
# Yeniden şekillendirme (Reshape / Shape)
dizi_16 = np.arange(16)
print("İlk Boyut Sayısı:", dizi_16.ndim)
dizi_16.shape = (4, 4)
print("\n4x4 Matris Hali:\n", dizi_16)
print("\nYeni Boyut Sayısı:", dizi_16.ndim)
```

```python
# Çok boyutlu diziyi tek boyuta indirme (Ravel)
dizi_cok = np.array([[2, 5, 7], [3, 6, 8]])
düz_dizi = dizi_cok.ravel()
print("Düzleştirilmiş Dizi:", düz_dizi)
```

5. Matematiksel ve İstatistiksel İşlemler
Diziler üzerinde toplu hesaplama yapmanızı sağlayan fonksiyonlar.

NumPy Matematiksel Fonksiyonları ve Görevleri
Aşağıda bu fonksiyonlar ve diğer bazı fonksiyonların görevleri listelenmiştir:

**Abs:** Bu fonksiyon yerine absolute() fonksiyonu da kullanılabilir. Bir dizinin her bir elemanının mutlak değerinin hesaplanmasında kullanılır.

**Fabs:** Bu fonksiyon da mutlak değer hesaplamak için kullanılır. Abs() fonksiyonundan farkı, mutlak değeri hesaplanan değer integer tipinde olsa bile sonucun float tipinde dönmesidir.

**Sqrt:** Bir dizinin her bir elemanının karekökünün hesaplanmasında kullanılır.

**Square:** Bir dizinin her bir elemanının karesinin hesaplanmasında kullanılır.

**Isnan:** Bir dizinin her bir elemanının içeriğinin boş olup olmadığı test edilir. Sonuç ise True ya da False olarak boolean tipinde geri döner.


**Exp:** Bir dizinin her bir elemanının üstel (exponential) değerinin hesaplanmasında kullanılır.

**Cos:** Bir dizinin her bir elemanının kosinüs değerinin hesaplanmasında kullanılır.

**Sign:** Bir dizinin her bir elemanının işaretini geri döndürür. Eğer dizi elemanı negatif ise -1, pozitif ise 1 ve sıfır ise 0 olarak döndürür.


**Ceil:** Bir dizinin her bir elemanının bir üst sayıya yuvarlanmasında kullanılır.  Örneğin; 0.53 → 1 ya da 10.1 → 11 gibi.

**Floor:** Bu fonksiyon, ceil() fonksiyonu gibi bir dizinin her bir elemanının yuvarlanmasında kullanılır. Bu fonksiyonun farkı ise bir üst sayıya değil alt sayıya yuvarlamasıdır.  Örneğin; 0.53 → 0 ya da 10.1 → 10 gibi.

```python
import numpy as np

# Örnek bir veri seti oluşturalım (Negatif, ondalıklı ve özel değerler içeren)
veri = np.array([-5.6, 10.1, 0, np.nan, 4])

print("Orijinal Veri Seti:", veri)

```

```python
# 1. Abs / Absolute: Mutlak değer (Negatifi pozitife çevirir) [cite: 295, 296]
print("Abs (Mutlak Değer):", np.abs(veri))
```

```python
# 2. Fabs: Her zaman float tipinde mutlak değer döndürür [cite: 297, 298]
# Not: nan değerleri korur
print("Fabs (Float Mutlak):", np.fabs([-5, 10, -3]))

```

```python
# 3. Sqrt: Karekök hesaplama [cite: 299]
# Not: Negatif sayıların karekökü 'nan' döner
print("Sqrt (Karekök):", np.sqrt([16, 25, 9]))
```

```python
# 4. Square: Kare alma [cite: 299]
print("Square (Karesi):", np.square([2, 3, 4]))
```

```python
# 5. Isnan: Boş/Tanımsız (NaN) kontrolü [cite: 299, 301]
print("Isnan (Boş mu?):", np.isnan(veri))
```

```python
# 6. Exp: Üstel değer hesaplama (e^x) [cite: 300, 303]
print("Exp (Üstel):", np.exp([1, 2]))
```

```python
# 7. Cos: Kosinüs hesaplama [cite: 304, 306]
print("Cos (Kosinüs):", np.cos([0, np.pi]))
```

```python
# 8. Sign: İşaret belirleme (-1, 0, 1) [cite: 305, 307]
print("Sign (İşaret):", np.sign([-10, 0, 25]))
```

```python
# 9. Ceil: Bir üst tam sayıya yuvarlama [cite: 308, 309]
print("Ceil (Üste Yuvarla):", np.ceil([0.53, 10.1]))
```

```python
# 10. Floor: Bir alt tam sayıya yuvarlama [cite: 310, 312]
print("Floor (Alta Yuvarla):", np.floor([0.53, 10.1]))
```

```python
d1 = np.array([4, 6, 8, 9])
d2 = np.array([2, 4, 3, 5])

print("Toplam:", d1 + d2)
print("\nÇarpım:", d1 * d2)
print("\nBölüm (Ondalık):", d1 / d2)
print("\nÜst (Kuvvet):", d1 ** d2)
```

```python
# Temel İstatistikler
veri = np.array([[1, 2.5, 6], [8, 20, 11]])
print("En Küçük Değer:", veri.min())
print("En Büyük Değer:", veri.max())
print("Standart Sapma:", veri.std())
```

```python
# Matematiksel Fonksiyonlar
print("-" * 30)
print(f"{'MATEMATİKSEL İŞLEMLER':^30}")
print("-" * 30)

print(f"🔹 Karekök Değerleri:\n{np.sqrt(veri)}\n")
print(f"🔹 Üste Yuvarlanmış (Ceil):\n{np.ceil(veri)}\n")
print(f"🔹 Eksik Veri (NaN) Kontrolü:\n{np.isnan(veri)}")
```

6. Dilimleme (Slicing) ve Birleştirme
Veri setinden belirli bölümleri seçme ve farklı dizileri bir araya getirme.

```python
# Tek boyutlu dilimleme
sira = np.array([5, 12, 4, -2, 15, 6, 8, 0])
print("1'den 5. indise kadar:", sira[1:5])
print("Diziyi ters çevir:", sira[::-1])
```

```python
# Orijinal diziyi koruyarak kopyalama
yeni_kopya = sira[2:4].copy()
sira[2] = 200
print("Kopya Dizi (Orijinaldeki değişimden etkilenmez):", yeni_kopya)
```

```python
# Çok boyutlu dilimleme (Satır ve Sütun seçimi)
matris = np.arange(12).reshape(3, 4)
print("0. Satırın Tamamı:", matris[0, :])
print("2. Sütunun Tamamı:", matris[:, 2])
```

```python
# Dizileri dikey olarak birleştirme (Concatenate)
v1 = np.full((2, 4), 0.0)
v2 = np.full((3, 4), 1.0)
birlesik = np.concatenate((v1, v2), axis=0)
print("Birleşik Dizi (Axis 0):\n", birlesik)
```

```python
# Diziyi ayırma (Split)
parcalar = np.split(np.arange(12.0), 4)
print("4 Parçaya Ayrılmış Dizi:", parcalar)

# Diziyi doğrudan böl ve döngüye sok
for i, parca in enumerate(parcalar, 1):
    print(f"Dizi {i} -> Array: {parca}")
```
